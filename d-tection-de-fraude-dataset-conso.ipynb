```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
#Chargement du dataframe et séparation des colonnes
df = pd.read_csv('data.csv')
labels = pd.DataFrame()
labels['FLAG'] = df['FLAG']

labels['CONS_NO'] = df['CONS_NO']

print(labels)

data = df.drop(['FLAG', 'CONS_NO'], axis=1)
data
```

```python
# suppression des doublons
di = data[data.duplicated()].index 
data = data.drop(di, axis=0)
labels = labels.drop(di, axis=0)
```

```python
#suppression des valeurs nulles

zi = data[(data.sum(axis=1) == 0)].index  
data = data.drop(zi, axis=0)
labels = labels.drop(zi, axis=0)
```

```python
# réindexation des colonnes 
data.columns = pd.to_datetime(data.columns)  
data = data.reindex(sorted(data.columns), axis=1)
cols = data.columns
```

```python
data.reset_index(inplace=True, drop=True)  
labels.reset_index(inplace=True, drop=True)
```

```python
#remplissage des valeurs NaN
data = data.interpolate(method='linear', limit=2,  
                        limit_direction='both', axis=0).fillna(0)
```

```python
data
```

Traitement des Outliers

```python
#traitement des outliers
for i in range(data.shape[0]):  
    m = data.loc[i].mean()
    st = data.loc[i].std()
    data.loc[i] = data.loc[i].mask(data.loc[i] > (m + 3 * st), other=m + 3 * st)
```

```python
vdata = data  #donnée de visualisaion avant de faire le Scaling
```

```python

scale = MinMaxScaler()
scaled = scale.fit_transform(data.values.T).T
mData = pd.DataFrame(data=scaled, columns=data.columns)
print(mData)
preData = pd.concat([labels, mData], axis=1, sort=False) 
print(preData)

```
